## Adding 3 new columns ( hashtags, mentions, urls )

```python
#importing regex
import re 
```

```python
#finding all the hashtags using lambda functions
df['hashtag'] = df['text'].apply(lambda x: re.findall(r"#[A-Za-z0-9_]+", x)) 
#removig # and [] from the hashtag column
df['hashtag'] = df['hashtag'].apply(lambda x : str(x).replace('[','').replace(']','').replace('#','').strip("'")) 
```

```python
df.head(25)
```

```python
#finding all the urls in the text
df['url'] = df['text'].apply(lambda x: re.findall(r"http[s]?://(?:[a-zA-Z]|[0-9]|[$-_@.&+]|[!*\(\),]|(?:%[0-9a-fA-F][0-9a-fA-F]))+", x))
df['url'] = df['url'].apply(lambda x : str(x).replace('[','').replace(']','').strip("'"))
```

```python
df
```

```python
#finding mentions in the text
df['mentions'] = df['text'].apply(lambda x: re.findall(r"@[A-Za-z0-9_]+", x))
```

```python
#removing @ and [] from the mentions column
df['mentions'] = df['mentions'].apply(lambda x : str(x).replace('[','').replace(']','').replace('@','').strip("'"))
```

```python
#creating a column for storing clean data
df['clean_text'] = df['text'].copy()
```

```python
df
```

```python
# removing hashtags from the clean_text columns
df['clean_text'] = df['clean_text'].apply(lambda x: re.sub(r"#[A-Za-z0-9_]+","", x))
```

```python
df
```

```python
#removing urls from the clean_text columns
df['clean_text'] = df['clean_text'].apply(lambda x: re.sub(r"http[s]?://(?:[a-zA-Z]|[0-9]|[$-_@.&+]|[!*\(\),]|(?:%[0-9a-fA-F][0-9a-fA-F]))+","", x))
```

```python
df
```

```python
#removing mentions from the clean_text 
df['clean_text'] = df['clean_text'].apply(lambda x: re.sub(r"@[A-Za-z0-9_]+","", x))
```

```python
df
```

```python
from string import punctuation
```

## Removing punctuations

```python
df['clean_text'] = df['clean_text'].apply(lambda x: re.sub(r'[' + punctuation.replace('@', '') + ']+',"", x))
```

```python
df
```

## Removing special characters

```python
df['clean_text'] = df['clean_text'].apply(lambda x: re.sub(r"\&\w*;","", x))
```

```python
df
```

```python
df.clean_text[81]
```

```python
df['clean_text'] = df['clean_text'].apply(lambda x: re.sub(r"/[1F601-1F64F]/u", r"", x)) #removing emojis from the text
df
```

## Adding space between potential words

```python
df['clean_text'] = df['clean_text'].apply(lambda x: re.sub(r"([A-Z])", r"\1", x))
df
```

## Removing extra spaces

```python
df['clean_text'] = df['clean_text'].str.strip()
```

```python
df
```

## Converting the text into lowercase.

```python
df['clean_text'] = df['clean_text'].str.lower()
```
